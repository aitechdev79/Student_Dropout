Prepare the feature set `X` by identifying and one-hot encoding all categorical variables (including those currently represented as integers but are categorical in nature) and ensuring numerical features are ready. ## Prepare Data for Modeling

### Subtask:
Prepare the feature set X by identifying and one-hot encoding all categorical variables (including those currently represented as integers but are categorical in nature) and ensuring numerical features are ready. Then, split the processed X and the target variable y into new training and testing sets for the subsequent machine learning models.


**Reasoning**:
To prepare the data for modeling, I need to identify both explicit and implicit categorical features, then apply one-hot encoding to these. Next, I will identify the truly numerical features and combine them with the one-hot encoded categorical features to form a complete feature set `X_processed`. Finally, I will split `X_processed` and the target `y` into training and testing sets.



```python
from sklearn.model_selection import train_test_split

# 1. Identify categorical columns (including those represented as integers)
categorical_features = [
    'Marital status', 'Application mode', 'Application order', 'Course',
    'Daytime/evening attendance', 'Previous qualification', 'Nacionality',
    'Mother\'s qualification', 'Father\'s qualification', 'Mother\'s occupation',
    'Father\'s occupation', 'Displaced', 'Educational special needs', 'Debtor',
    'Tuition fees up to date', 'Gender', 'Scholarship holder', 'International'
]

# Ensure these columns are treated as categorical for get_dummies
# Convert them to 'category' dtype temporarily if they are int, to prevent issues if there are many unique values
for col in categorical_features:
    if col in df.columns:
        df[col] = df[col].astype('category')

# 2. Identify numerical columns
# Exclude 'Target', 'Target_encoded', and the identified categorical features
all_columns_except_target = df.drop(columns=['Target', 'Target_encoded'], errors='ignore').columns
numerical_features = [col for col in all_columns_except_target if col not in categorical_features]

# 3. Apply one-hot encoding to categorical features
X_categorical_processed = pd.get_dummies(df[categorical_features], drop_first=True)

# 4. Combine numerical and processed categorical features
X_processed = pd.concat([df[numerical_features], X_categorical_processed], axis=1)

# Ensure y is the encoded target variable
y = df['Target_encoded']

# 5. Split the data into training and testing sets
X_train_processed, X_test_processed, y_train_processed, y_test_processed = train_test_split(
    X_processed, y, test_size=0.3, random_state=42, stratify=y
)

# 6. Print the shapes
print(f"Shape of X_processed: {X_processed.shape}")
print(f"Shape of X_train_processed: {X_train_processed.shape}")
print(f"Shape of X_test_processed: {X_test_processed.shape}")
print(f"Shape of y_train_processed: {y_train_processed.shape}")
print(f"Shape of y_test_processed: {y_test_processed.shape}")
```

## Define Models for Evaluation

### Subtask:
Initialize the RandomForestClassifier, GradientBoostingClassifier, and XGBoostClassifier models. Set `random_state` for reproducibility and specify any initial parameters for these models.


**Reasoning**:
To initialize the required models, I will import them from their respective libraries and then instantiate each one with the specified parameters to ensure reproducibility and correct configuration.



```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Initialize RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
print("RandomForestClassifier initialized.")

# Initialize GradientBoostingClassifier
gbc_model = GradientBoostingClassifier(random_state=42, n_estimators=100)
print("GradientBoostingClassifier initialized.")

# Initialize XGBoostClassifier
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
print("XGBClassifier initialized.")
```

## Implement Cross-Validation for Each Model

### Subtask:
Utilize K-Fold Cross-Validation to evaluate each model's performance on the training data. This will provide a more reliable estimate of model performance compared to a single train-test split. We will track average performance metrics (e.g., accuracy, f1-score).


**Reasoning**:
To implement K-Fold Cross-Validation, I need to import the necessary tools from scikit-learn, set up the cross-validation strategy, and then iterate through each model to calculate and display its cross-validation scores.



```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Define the number of folds for cross-validation
k_folds = 5

# Initialize StratifiedKFold for maintaining class distribution across folds
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

models = {
    "RandomForestClassifier": rf_model,
    "GradientBoostingClassifier": gbc_model,
    "XGBClassifier": xgb_model
}

results = {}

print(f"Performing {k_folds}-Fold Stratified Cross-Validation...")

for name, model in models.items():
    print(f"\nEvaluating {name}...")
    # Perform cross-validation
    # Using 'accuracy' as scoring metric, can be changed to 'f1_weighted' for imbalanced datasets if needed
    scores = cross_val_score(model, X_train_processed, y_train_processed, cv=skf, scoring='accuracy', n_jobs=-1)

    # Store and print results
    results[name] = {
        "mean_accuracy": scores.mean(),
        "std_accuracy": scores.std()
    }
    print(f"  Mean Accuracy: {scores.mean():.4f}")
    print(f"  Standard Deviation of Accuracy: {scores.std():.4f}")

print("\nCross-validation complete.")
```

## Hyperparameter Tuning (Optional but Recommended)

### Subtask:
Perform hyperparameter tuning for RandomForestClassifier, GradientBoostingClassifier, and XGBoostClassifier using RandomizedSearchCV to find optimal parameters.


**Reasoning**:
To perform hyperparameter tuning, I will import RandomizedSearchCV, define parameter distributions for each model, and then iterate through RandomForestClassifier, GradientBoostingClassifier, and XGBoostClassifier, applying RandomizedSearchCV to find and print their optimal parameters and scores.



```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

print("Starting Hyperparameter Tuning with RandomizedSearchCV...")

# --- 1. RandomForestClassifier Tuning ---
print("\n--- Tuning RandomForestClassifier ---")
param_dist_rf = {
    'n_estimators': randint(50, 200),
    'max_features': ['sqrt', 'log2', None],
    'max_depth': randint(10, 50),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10)
}

random_search_rf = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dist_rf,
                                      n_iter=50, cv=skf, scoring='accuracy', random_state=42, n_jobs=-1, verbose=1)
random_search_rf.fit(X_train_processed, y_train_processed)

print("Best parameters for RandomForestClassifier:", random_search_rf.best_params_)
print("Best accuracy for RandomForestClassifier:", random_search_rf.best_score_)

# --- 2. GradientBoostingClassifier Tuning ---
print("\n--- Tuning GradientBoostingClassifier ---")
param_dist_gbc = {
    'n_estimators': randint(50, 200),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.6, 0.4) # between 0.6 and 1.0
}

random_search_gbc = RandomizedSearchCV(estimator=gbc_model, param_distributions=param_dist_gbc,
                                       n_iter=50, cv=skf, scoring='accuracy', random_state=42, n_jobs=-1, verbose=1)
random_search_gbc.fit(X_train_processed, y_train_processed)

print("Best parameters for GradientBoostingClassifier:", random_search_gbc.best_params_)
print("Best accuracy for GradientBoostingClassifier:", random_search_gbc.best_score_)

# --- 3. XGBoostClassifier Tuning ---
print("\n--- Tuning XGBoostClassifier ---")
# Note: xgb_model was already initialized with use_label_encoder=False and eval_metric='mlogloss'
param_dist_xgb = {
    'n_estimators': randint(50, 200),
    'learning_rate': uniform(0.01, 0.2),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4)
}

random_search_xgb = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_dist_xgb,
                                      n_iter=50, cv=skf, scoring='accuracy', random_state=42, n_jobs=-1, verbose=1)
random_search_xgb.fit(X_train_processed, y_train_processed)

print("Best parameters for XGBoostClassifier:", random_search_xgb.best_params_)
print("Best accuracy for XGBoostClassifier:", random_search_xgb.best_score_)

print("\nHyperparameter tuning complete.")
```
